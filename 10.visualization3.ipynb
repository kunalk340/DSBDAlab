Download the Iris flower dataset or any other dataset into a DataFrame. (e.g.,
https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and give the
inference as: 
1. List down the features and their types (e.g., numeric, nominal) available in the dataset. 
2. Create a histogram for each feature in the dataset to illustrate the feature distributions. 
3. Create a boxplot for each feature in the dataset. 
4. Compare distributions and identify outliers.

```python
# libraries &amp; dataset
import seaborn as sns
import matplotlib.pyplot as plt

# set a grey background 
sns.set(style="darkgrid")
df = sns.load_dataset("iris")

# list all the features and their types
print(df.dtypes)
```

```python
figure = plt.figure(figsize=(15, 7))
sns.histplot(data=df, x="sepal_length", hue="species")
plt.show()

figure = plt.figure(figsize=(15, 7))
sns.histplot(data=df, x="sepal_length")
plt.show()
```

```python
figure = plt.figure(figsize=(15, 7))
sns.histplot(data=df, x="sepal_width",hue="species")
plt.show()

figure = plt.figure(figsize=(15, 7))
sns.histplot(data=df, x="sepal_width")
plt.show()
```

```python
figure = plt.figure(figsize=(15, 7))
sns.histplot(data=df, x="petal_length",hue="species")
plt.show()

figure = plt.figure(figsize=(15, 7))
sns.histplot(data=df, x="petal_length")
plt.show()
```

```python
figure = plt.figure(figsize=(15, 7))
sns.histplot(data=df, x="petal_width",hue="species")
plt.show()

figure = plt.figure(figsize=(15, 7))
sns.histplot(data=df, x="petal_width")
plt.show()
```

```python
figure = plt.figure(figsize=(15, 7))
sns.boxplot(data=df, x='species', y="sepal_length",hue="species")
plt.show()
```

```python
figure = plt.figure(figsize=(15, 7))
sns.boxplot(data=df, x='species', y="sepal_width",hue="species")
plt.show()
```

```python
figure = plt.figure(figsize=(15, 7))
sns.boxplot(data=df, x='species', y="petal_length",hue="species")
plt.show()
```

```python
figure = plt.figure(figsize=(15, 7))
sns.boxplot(data=df, x='species', y="petal_width",hue="species")
plt.show()
```
