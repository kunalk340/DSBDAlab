```python
# Print the number of missing values if any
missing_values=df.isnull().sum()
print(missing_values)
```

```python
miss_val_perc = (missing_values * 100) / len(df)
print(miss_val_perc)
```

```python
# displaying only the numerical data 

numerical_data = df.select_dtypes(include=np.number)
print(numerical_data)
```

```python
# as calories cannot be negative value the negative values are converted using abs() function

df["Calories"]=df["Calories"].abs()
df
```

```python
#replacing the missing values in Calories column by the mean method

df['Calories']=df['Calories'].replace(np.NaN,df['Calories'].mean())
df
```

```python
#again calculating the no. of NaN values in calories column after handling

df['Calories'].isnull().sum()
```

```python
df.dtypes
```

```python
#converting one data type to another, here 'Date' object type is converted to datetime type

df['Date'] = pd.to_datetime(df['Date'])
print(df.to_string())
df.dtypes
```

```python
# as row 22 has a missing value for date, we drop that particular row

df.dropna(subset=['Date'], inplace = True)
df
```

```python
#converting the data type of calories from float to int

df_int = df
df_int['Calories'] = df_int['Calories'].astype('int')
print(df_int)
df_int.dtypes
```
